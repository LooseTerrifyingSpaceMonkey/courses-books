### Pandas Exercise Donut Dataset
- Generate random data x1, x2, y into a dataframe
- Add the additional columns
- Plot the donut
- Save to a CSV without header row, index columns

```
sample = (np.random.randn(1000, 2) * 2) - 1
```

```
donut_df = pd.DataFrame(data=sample, columns=['X', 'Y'])
```

```
donut_df.head()
```

```
donut_df.plot.scatter('X', 'Y')
```

```
# df['year'] = df.apply(date_to_year, axis=1)
donut_df['X2'] = donut_df.apply(lambda row: row['X'] ** 2, axis=1)
```

```
donut_df.head()
```

```
donut_df['Y2'] = donut_df.apply(lambda row: row['Y'] ** 2, axis=1)
```

```
donut_df['XY'] = donut_df.apply(lambda row: row['X'] * row['Y'], axis=1)
```

```
donut_df.head()
```

```
donut_df['Z'] = donut_df.apply(lambda row: 0 if (row['X'] > 0 and row['Y'] > 0) or (row['X'] <= 0 and row['Y'] <= 0) else 1, axis=1)
```

```
donut_df.head(10)
```

```
donut_df.plot.scatter(x='X2', y='Y2', color='Z', colormap='viridis')
```

```
donut_df.to_csv('output/pandas_output.csv', index=False, header=False)
```
